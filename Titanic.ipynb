```python
import pandas as pd
```

```python
# Cargar el conjunto de datos del Titanic en un DataFrame llamado 'df'
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Ver el porcentaje de supervivencia agrupado por género
df.groupby('sex')['survived'].mean()
```

```python
# Ver el porcentaje de supervivencia agrupado por clase
df.groupby('pclass')['survived'].mean()
```

```python
# Calcular la edad promedio de todos los pasajeros
df['age'].mean()
```

```python
# Rellenar los valores NaN (vacíos) en la columna 'age' con la edad promedio
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Crear una gráfica de barras (countplot)
sns.countplot(data=df, x='pclass', hue='survived')
```

```python
# Crear una gráfica de barras para supervivencia por género
sns.countplot(data=df, x='sex', hue='survived')
```
